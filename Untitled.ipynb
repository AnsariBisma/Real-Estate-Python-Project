```python
#Calculate the percentage of missing value in each columns
percentage_train = df_train.isnull().sum()/len(df_train)*100

percentage_train
```

```python
percentage_test = df_test.isnull().sum()/len(df_test)*100
percentage_test
```

```python
#lets create the data frame for missing value in percentage 

percentage_train = pd.DataFrame(percentage_train, columns = ['Percentage of missing value'])
percentage_train.sort_values(by = ['Percentage of missing value'], inplace = True, ascending = False)
percentage_train
```

```python
percentage_test = pd.DataFrame(percentage_test, columns = ['Percentage of missing values'])
percentage_test.sort_values(by = ['Percentage of missing values'], inplace = True , ascending = False)
percentage_test
```

```python
df_train.drop(columns = ['BLOCKID','SUMLEVEL'], inplace = True)
df_test.drop(columns = ['BLOCKID','SUMLEVEL'], inplace = True)
```

```python
# lets find out all missing values columns with creating loop
missing_train_cols = []
for col in df_train.columns:
    if df_train[col].isnull().sum() !=0:
        missing_train_cols.append (col)
        
print(missing_train_cols)
        

```

```python
missing_test_cols = []
for col in df_test.columns:
    if df_test[col].isnull().sum() != 0:
        missing_test_cols.append (col)

print(missing_test_cols)
```

```python
# lets treat the missing values by mean bcoz all columns are numeric , here we are using loops bcoz we have lots of missing 
# value columns not possible to do sepratly hence using loop

for col in df_train.columns:
    if col in (missing_train_cols):
        df_train[col].replace(np.nan, df_train[col].mean(), inplace = True)
        


```

```python
for col in df_test.columns:
    if col in (missing_test_cols):
        df_test[col].replace(np.nan, df_test[col].mean(), inplace = True)
```
